```python
import numpy as np
from astropy.io import fits
from astropy.table import Table
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
archivos= ['DataDESI_76.fits', 'DataDESI_152.fits'
           , 'DataDESI_213.fits', 'DataDESI_284.fits', 'DataDESI_351.fits',
           'DataDESI_438.fits', 'DataDESI_530.fits', 'DataDESI_606.fits', 'DataDESI_690.fits', 'DataDESI_752.fits']
```

```python
pred = fits.open('SPECTYPE_SPASS.fits')
```

```python
RA = np.array([])
DEC = np.array([])
Z = np.array([])

for h in range(len(archivos)):
    espc = fits.open(archivos[h]) #open file    
    # Obtener la clase espectral y corrimiento para cada espectro
    TARGET_RA = Table.read(espc, hdu=1)['TARGET_RA'].data
    TARGET_DEC = Table.read(espc, hdu=1)['TARGET_DEC'].data
    corrimiento = Table.read(espc, hdu=1)['Z'].data
    
    Z = np.append(Z, corrimiento)
    RA = np.append(RA, TARGET_RA)
    DEC = np.append(DEC, TARGET_DEC)
    
```

```python
STYPE = np.array([])
for i in range(1,11):
    type = Table.read(pred, hdu=i)['SPECTYPE'].data
    STYPE = np.append(STYPE, type)
    
```

```python
#STYPE, RA, DEC, Z
```

```python
#STYPE = np.array(['STAR', 'STAR', 'GALAXY', 'QSO', 'GALAXY', 'QSO'])
```

```python
# Encontrar los Ã­ndices de 'QSO'
indices_qso = np.where(STYPE == 'QSO')[0]
indices_gal = np.where(STYPE == 'GALAXY')[0]
indices_star = np.where(STYPE == 'STAR')[0]
```

```python
# Filtrar los valores correspondientes a 'QSO'
Z_qso = Z[indices_qso]
RA_qso = RA[indices_qso]
DEC_qso = DEC[indices_qso]

# Filtrar los valores correspondientes a 'GALAXY'
Z_gal = Z[indices_gal]
RA_gal = RA[indices_gal]
DEC_gal = DEC[indices_gal]

#Filtrar los valores correspondientes a 'STAR'
Z_star = Z[indices_star]
RA_star = RA[indices_star]
DEC_star = DEC[indices_star]
```

```python
# Graficar Z vs RA solo para los valores correspondientes a 'QSO'
plt.scatter(RA_qso, Z_qso, color='red', label='QSO')
plt.xlabel('RA')
plt.ylabel('Z')
plt.title('Z vs RA')
plt.grid(True)
plt.legend()
#plt.savefig('RAvsZ_qso')
plt.show()
plt.close()

# Graficar Z vs RA solo para los valores correspondientes a 'GALAXY'
plt.scatter(RA_gal, Z_gal, color='blue', label='GALAXY')
plt.xlabel('RA')
plt.ylabel('Z')
plt.title('Z vs RA')
plt.grid(True)
plt.legend()
#plt.savefig('RAvsZ_gal')
plt.show()
plt.close()

# Graficar Z vs RA solo para los valores correspondientes a STAR'
plt.scatter(RA_star, Z_star, color='green', label='STAR')
plt.xlabel('RA')
plt.ylabel('Z')
plt.title('Z vs RA')
plt.grid(True)
plt.legend()
plt.savefig('RAvsZ_star')
plt.show()
plt.close()

# Graficar Z vs RA
plt.scatter(RA_qso, Z_qso, color='red', label='QSO')
plt.scatter(RA_gal, Z_gal, color='blue', label='GALAXY')
plt.scatter(RA_star, Z_star, color='green', label='STAR')
plt.xlabel('RA')
plt.ylabel('Z')
plt.title('Z vs RA')
plt.grid(True)
plt.legend()
#plt.savefig('RAvsZ_all')
plt.show()
plt.close()
```

```python
# Graficar Z vs RA solo para los valores correspondientes a 'QSO'
plt.scatter(DEC_qso, Z_qso, color='red', label='QSO')
plt.xlabel('DEC')
plt.ylabel('Z')
plt.title('Z vs DEC')
plt.grid(True)
plt.legend()
plt.savefig('DECvsZ_qso')
plt.show()
plt.close()

# Graficar Z vs RA solo para los valores correspondientes a 'GALAXY'
plt.scatter(DEC_gal, Z_gal, color='blue', label='GALAXY')
plt.xlabel('DEC')
plt.ylabel('Z')
plt.title('Z vs DEC')
plt.grid(True)
plt.legend()
plt.savefig('DECvsZ_gal')
plt.show()
plt.close()

# Graficar Z vs RA solo para los valores correspondientes a STAR'
plt.scatter(DEC_star, Z_star, color='green', label='STAR')
plt.xlabel('DEC')
plt.ylabel('Z')
plt.title('Z vs DEC')
plt.grid(True)
plt.legend()
plt.savefig('DECvsZ_star')
plt.show()
plt.close()

# Graficar Z vs RA
plt.scatter(DEC_qso, Z_qso, color='red', label='QSO')
plt.scatter(DEC_gal, Z_gal, color='blue', label='GALAXY')
plt.scatter(DEC_star, Z_star, color='green', label='STAR')
plt.xlabel('DEC')
plt.ylabel('Z')
plt.title('Z vs DEC')
plt.grid(True)
plt.legend()
#plt.savefig('RAvsZ_all')
plt.show()
plt.close()
```

```python
# Graficar Z vs RA solo para los valores correspondientes a 'QSO'
plt.scatter(RA_qso, DEC_qso, color='red', label='QSO')
plt.xlabel('RA')
plt.ylabel('DEC')
plt.title('DEC vs RA')
plt.grid(True)
plt.legend()
plt.savefig('RAvsDEC_qso')
plt.show()
plt.close()

# Graficar Z vs RA solo para los valores correspondientes a 'GALAXY'
plt.scatter(RA_gal, DEC_gal, color='blue', label='GALAXY')
plt.xlabel('RA')
plt.ylabel('DEC')
plt.title('DEC vs RA')
plt.grid(True)
plt.legend()
plt.savefig('RAvsDEC_gal')
plt.show()
plt.close()

# Graficar Z vs RA solo para los valores correspondientes a STAR'
plt.scatter(RA_star, DEC_star, color='green', label='STAR')
plt.xlabel('RA')
plt.ylabel('DEC')
plt.title('DEC vs RA')
plt.grid(True)
plt.legend()
plt.savefig('RAvsDEC_star')
plt.show()
plt.close()

# Graficar Z vs RA
plt.scatter(RA_qso, DEC_qso, color='red', label='QSO')
plt.scatter(RA_gal, DEC_gal, color='blue', label='GALAXY')
plt.scatter(RA_star, DEC_star, color='green', label='STAR')
plt.xlabel('RA')
plt.ylabel('DEC')
plt.title('DEC vs RA')
plt.grid(True)
plt.legend()
#plt.savefig('RAvsZ_all')
plt.show()
plt.close()
```

```python
# Crear figura con tamaÃ±o aumentado
fig = plt.figure(figsize=(12, 8))  # Ajusta los valores de figsize segÃºn tus necesidades
ax = fig.add_subplot(111, projection='3d')

# Definir el espaciado entre los ticks en los ejes RA y DEC
ax.set_xticks(np.arange(min(RA), max(RA) + 1, 20))  # Define los ticks cada 20 unidades en RA
ax.set_yticks(np.arange(min(DEC), max(DEC) + 1, 10)) # Define los ticks cada 10 unidades en DEC
ax.set_zticks(np.arange(min(Z), max(Z) + 0.5, 0.5))  # Define los ticks cada 0.5 unidades en Z

# Graficar puntos
ax.scatter(RA_qso, DEC_qso, Z_qso, color='red', label='QSO', alpha=0.5)
ax.scatter(RA_gal, DEC_gal, Z_gal, color='blue', label='GALAXY', alpha=0.5)
ax.scatter(RA_star, DEC_star, Z_star, color='green', label='STAR', alpha=0.5)

# Etiquetas y tÃ­tulo
ax.set_xlabel('RA')
ax.set_ylabel('DEC')
ax.set_zlabel('Z')
ax.set_title('Z vs RA vs DEC')
ax.legend()

# Guardar la figura antes de mostrarla
plt.savefig('3D_desi.png', bbox_inches='tight')
# Mostrar y cerrar
plt.show()
plt.close()
```
